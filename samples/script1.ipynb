```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000)

plt.axvline(0, color='black', lw=1)
plt.plot(x, np.sin(x)/x)
plt.show()
```
